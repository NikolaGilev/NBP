```python
movies_df['genres'].replace('['']','None', inplace=True)

import json
# genre_row is a list of jsons represented in a string
all_genre_rows = []
for index,genre_row in enumerate(movies_df['genres']):
    if genre_row != 'None':
        # we need to split the string like so
        data_string = genre_row[1:-1].split("},")
        # we need to add back '}' to the elements we split and ignore the last one
        genre_row_elements = [ds + '}' for ds in data_string[:-1]]
        # we add the last one normaly
        genre_row_elements.append(data_string[-1])


        all_row_elements = []
        for element in genre_row_elements:
            element_fixed = element.replace("'", "\"").replace("None", "null")
            json_element = json.loads(element_fixed)
            json_element['movie_id'] = movies_df.at[index, 'id']
            all_row_elements.append(json_element)
        all_genre_rows.append(all_row_elements)
    else:
        all_genre_rows.append([np.nan])
all_genre_rows
```

```python
len(movies_df["genres"])
```

```python
from itertools import chain
flattened_genre_list = list(chain(*all_genre_rows))
flattened_genre_list
```

```python
df_genre = pd.DataFrame([item for item in flattened_genre_list if isinstance(item, dict)])
df_genre
```

```python
df_genre.to_csv('./csvs/genres.csv')
```
